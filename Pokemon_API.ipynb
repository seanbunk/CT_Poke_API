```python
# POKEMON API
# Retrieve a list of Pokemon's name, abilities, types, and weight from the PokeAPI
#  https://pokeapi.co/

# Make 20 Pokemon
import requests, json
```

```python
pokeDex = ['bulbasaur','ivysaur','blastoise','abra','kadabra', 'charmander','charmeleon', 
                 'charizard','squirtle','wartortle','haunter','gengar' 'caterpie', 
                 'eevee','mewtwo','mew', 'togepi','articuno','machop','scyther',
                 'golem','dragonite']

for name in pokeDex:
    url = f'https://pokeapi.co/api/v2/pokemon/{name}'
    response = requests.get(url)

    if response.ok:
        pokeDex = response.json()
        abilities = [ability['ability']['name'] for ability in pokeDex['abilities']]
        types = [type['type']['name'] for type in pokeDex['types']]
        weight = pokeDex['weight']

        print(f"Name: {name}")
        print(f"Abilities: {', '.join(abilities)}")
        print(f"Types: {', '.join(types)}")
        print(f"Weight: {weight}\n")
```
